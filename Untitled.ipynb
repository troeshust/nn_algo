```python
from src import MLP
import random
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
net = MLP(1, [4, 1])
```

```python
net.parameters()
```

```python
BATCH_SIZE = 315
STEPS = 1000


def eval(x):
    return net([x / 100.]).data * 100


def get_batch(size=BATCH_SIZE):
    X = [[random.randint(-200, 200) / 100.] for _ in range(size)]
    y = [x[0] for x in X]
    return X, y


def compute_accuracy():
    cnt = 0
    for x in range(1, 101):
        pred = round(eval(x))
        if pred == x:
            cnt += 1
    return cnt


print_next_plot_in = 1
for step in range(STEPS):
    X, y_trues = get_batch()
    y_preds = list(map(net, X))

    # L2 regularization
    alpha = 1e-4
    reg_loss = alpha * sum((p*p for p in net.parameters()))

    data_loss = sum([(y_true - y_pred) ** 2 for y_true, y_pred in zip(y_trues, y_preds)]) / BATCH_SIZE
    loss = data_loss + reg_loss

    net.zero_grad()
    loss.backward()
    lr = 0.1
    for param in net.parameters():
        param.data -= lr * param.grad
    
    if not step % 100:
        print(f'step {step}: loss {loss}, acc {compute_accuracy()}%')
    
    def print_plot():
        x = [i for i in np.arange(-300, 300, 0.2)]
        y = [e for e in x]
        pred = [eval(e) for e in x]
        fig = plt.figure()
        plt.plot(x, y)
        plt.plot(x, pred)
        plt.show()
    
    if step == print_next_plot_in:
        print_plot()
        print_next_plot_in = int(input()) + step
```

```python
def print_params():
    print('[' + ', '.join([str(p.data) for p in net.parameters()]) + ']')
```
